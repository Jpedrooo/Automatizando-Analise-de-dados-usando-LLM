```python
import numpy as np
import pandas as pd

df_reviwes = pd.read_csv("/content/reviews.csv")

```

```python
coluna_reviews = df_reviwes['reviewText']
coluna_reviews
```

```python


from google import genai
from google.colab import userdata



client = genai.Client(api_key=userdata.get('GEMINI_API_KEY_ALURA'))

lista_de_analises_de_sentimentos = []


for review_numero, resenha in enumerate(coluna_reviews):
    resposta = client.models.generate_content(
        model="gemini-2.0-flash",
        contents=f"""você ira analisar a resenha que eu te mandarei abaixo e rotornar com uma analise de sentimento.
Você deve responder apenas com uma das seguintes palavras: 'positiva', 'negativa', 'neutra'.
indicando o sentimento relativo aquela resenha espesifica.

exemplo:
'gostei do produto' = positiva
'nao achei nada de mais' = neutra
'odiei' = negativa

segue a resenha a ser analisada: {resenha}"""
    )

    lista_de_analises_de_sentimentos.append(resposta.text)

print(f"Resenha {review_numero}: {resenha} -> Sentimento: {resposta.text}")


df_reviwes["Analise de sentimentos"] = lista_de_analises_de_sentimentos

df_reviwes
```
